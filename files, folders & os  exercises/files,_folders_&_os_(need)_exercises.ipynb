```python
import os
import glob
import shutil
file_name=[]
for name in glob.glob('os-collaboration-exercise-data/**', recursive=True):
    file_name.append(os.path.split(name)[1])
print(file_name)
file_name=file_name[1:len(file_name):1]
print(file_name)
cities=['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for a in file_name:
    print(a)
    index1=a.index('_')
    index2=a.index('.')
    substr=a[index1+1:index2:1]
    for c in cities:
        if substr==c.lower():
            current=os.path.join('os-collaboration-exercise-data', a)
            dest=os.path.join('os-collaboration-exercise-data', c, a)
            shutil.copy(current, dest)
            break


```

Note on handling large-scale file modifications after this exercise: 

1. We should ideally first try to design a code, or use an AI to figure common strings/patterns that exist in the file name itself. Consequently we should organize the files accordingly in order to ensure that we can manipulate the data better later on. 
2. If there are files having multiple types of extensions, we should also parse through, and organize to make different directories for each type of extension - this will allow us to design code that is specific to parse through and correctly interpret the way in which data in stored in each file type, thus preventing data corruption. 
3. We should also ideally create a smaller representative subset of the larger data set, so that we can review, modify and test our code on the smaller subset, before trying to edit the main files (this is because changes to filenames, associated modifications, etc. cannot be reverted, so they need to be done carefully and systematically after testing).
