```python
%%timeit -r 1 -n 1
#exercise1
for i in range(1,10):
    if(i%2!=0):
        print((i**2))
    else:
        continue
```

```python
#exercise2
[(num**2) for num in range(1,10) if num%2!=0]
```

```python
#exercise
#for exercise1 - 515 µs ± 0 ns per loop (mean ± std. dev. of 1 run, 1 loop each) 
#for exercise2 - 6.3 µs ± 0 ns per loop (mean ± std. dev. of 1 run, 1 loop each) 
# hence list comprehension is faster than a for loop
```

```python
#exercise4
ini=100 
copy=ini
year=0  
while(copy>=(ini/2)):
    if(copy==(ini/2) or copy<(ini/2)):
        break
    year+=1
    copy=copy-(0.05*copy)
print("years taken to reduce to half-size=", year)
```
