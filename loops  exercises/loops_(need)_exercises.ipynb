```python
#exercise1
temperatures_celsius = [
    0, 5, 10, 15, 20, 25,
    30, 35, 40, 45, 50
]
for i in range(len(temperatures_celsius)):
    temp_fahr=(temperatures_celsius[i]*(9/5))+32
    print("temp in celsius=", temperatures_celsius[i])
    print(" converted temp in fahrenheit=", temp_fahr)

```

```python
#trial
for letter in ['A', 'B', 'C']:
    for number in [1, 2, 3]:
        print(f'{letter}{number}', end='\t')
    print('\n')
```

```python
#trial using single loop - creating a list of tuples using list comprehension
letters = ['A', 'B', 'C']
numbers = [1, 2, 3]

combined = [(letter, number) for letter in letters for number in numbers]

for pair in combined:
    print(f'{pair[0]}{pair[1]}', end='\t')
    if pair[1] == numbers[-1]:
        print('\n')
```
