## 1.1 Keyboard Shortcuts

```Stuff needed for Jupyter survival (Windows 11 ed.):```
|Action |Shortcut   |
|:-----:|:----------:|
|Run cell|```CTRL+ENTER```|
|Run and move to the next|```SHIFT+ENTER```|
|Convert cell to code cell|```ESC+Y```|
|Convert to Markdown|```ESC+M```|
|Create new cell|```ESC+A``` (below) and ```ESC+B``` (above)
|Copy cell|```ESC+C```|
|Paste cell|```ESC+V```|
|Merge cell|```SHIFT+M```|
|Delete cell (careful with this one!)|```ESC+D+D```|
|Show shortcuts|```SHIFT+H```| 




### Things to note

- Cells can be selected using the ```SHIFT``` and the up/down arrow keys simultaneously. 
- Careful with the `DELETE` command! 
- Shortcuts on Jupyter can be viewed and set by typing `ESC+H`

## 1.2 Shell commands

'Terminal' or shell commands can be run using Jupyter, by simply adding a ```!``` ahead of the command. 

For example, the *c*urrent working *d*irectory of the `.ipynb` file is found using the following command:
```
    !cd
```

The command ofr the list of all files in that particular working *dir*ectly:
```
    !dir
```

# 2 Shell Magic

The following in-built or `magic` commands can be used:
- `%timeit` - this command, at the beginning of a particular code line gives the user/coder statistics on the run time of that particular block of code. 
- `%%timeit` - same as above, but for a cell.
