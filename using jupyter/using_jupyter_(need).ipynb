## 4.4 Displaying Code

Inline code is incorporated in a Markdown cell using backticks ```('`')``` as follows: `print("hello world!")`

A block of code can be printed as follows: 
```python
a=3
b=4
c=a+b
print(c)
```


## 4.5 Links

Quite a trivial exercise - simply enclose the link (including the _https://_ portion) in the following brackets:
```'[]()'``` - this syntax allows us to create a hyper-link, so that the link inside the round brackets is hyperlinked and rendered as the text inside the square brackets. 

## 4.6 Images

Yet another trivial exercise - we use the same syntax as that of including links, only with an exclamation mark in the beginning of the statement as follows: ```'![](ABCD)'```

An *important* note (esp for Windows users): ensure that if you're including a local image, it's in the current working directory (i.e. the same as the directory of the ```.ipynb``` file)

## 4.9 Equations

The syntax for mathematical equations is essentially LaTEX. 
- Inline: $\sqrt{b^2-4ac}$
- Block of code: 
    $$
    x = \dfrac{-b \pm \sqrt{b^2-4ac}}{2a}
    $$

- Helpful commands can be found at the documentation for [Overleaf](https://www.overleaf.com/learn/latex/Mathematical_expressions).

- Here are some useful commands: 

|   |Input  |Rendered |
|:--:|:--:|:-----------:|
|Fraction|```$\dfrac{y}{x}$```|$\dfrac{y}{x}$|
|Subscript|```$x_{a}```|$x_{a}$|
|Power|```$x^{(y+z)}$```|$x^{(y+z)}$|
|Squareroot|```$\sqrt{a+b+c}$```|$\sqrt{(a+b+c)}$|
|Summation|```$\sum_{n=1}^{n=\infty} x_n$```|$\sum_{n=1}^{n=\infty} x_n$|
|Integral|```$\int_{n=1}^{n=\infty} f(x)dx$```|$\int_{n=1}^{n=\infty} f(x)dx$|
|Not Equal|```$a \ne b$```|$a \ne b$|
|Less than|```$a \lt b$```|$a \lt b$| 
|Less than or eq. to|```$a \leq b$```|$a \leq b$| 
|Greater than|```$a \gt b$```|$a \gt b$| 
|Greater than or eq. to|```$a \geq b$```|$a \geq b$| 
|Greek letters|```$alpha, \beta, \gamma, \pi, \lambda$```|$alpha, \beta, \gamma, \pi, \lambda$
