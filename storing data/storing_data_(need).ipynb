# What to expect in this chapter

This chapter will be discussing the three primary ways of storing different forms of data on Python: 
1. Python lists - a one-dimensional way of storing data, that can store different data types
2. NumPy Arrays - multi-dimensional way of storing data, that can store only one data type; NumPy arrays are very good for mathematical computations - as we will see later in this section
3. Dictionaries - can store two sets of related data, where one set is the 'key' to an associated 'value'

## 2.3 Comparison of mathematical operations

|Operation|List|Array|
|:---:|:---:|:---:|
|Adding a scalar number|Syntax error|numpy.array([])+10 (each element will be added by 10)
|Adding another list|list1 + list2 --> results in concatenation|results in the mathematical addition of corresponding elements|
|Multiplying by a scalar number (n)|results in concatenated list where the original list is repeated n times|each element of the array is multiplied by n|
|Squaring|Syntax error|each element of the array is squared|
|Using relational operators|Most relational operators do not work, `==` does not result in an error, but it makes no logical sense|Compares each element of the array with the relational operator
|Mathematical functions/attributes|Only the base Python functions will work on the whole list - more functions can be used by importing the `math` package|operations such as sum, max, min, mean, standard deviation can be called by using the related functions|

Notes about the `NumPy` arrays:
1. They are of **fixed size** - this is because they are allocated a fixed amount of memory on creation - thus allowing efficient memory management and operations on the array (more valid for large datasets). 
2. The values in it must have **uniform types** - this ensures that operations on it can performed efficiently. This also improves memory efficiency and allocation by ensuring that each element occupies the same amount of memory space (allowing overall performance gains).
