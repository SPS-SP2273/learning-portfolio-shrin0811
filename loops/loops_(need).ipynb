# 1 The for iterator

1. One of the most important and basic looping structures used.
2. Sequential loop, with a specified start and end value. 
3. Three ways of implementing a `for` loop:
    1. passing through a pre-defined list
    2. using the `enumerate()` function
    3. using the `range()` function

## 1.1 for with a list

The syntax for this system of working through a list is: `for i in list_name`:
1. Assigns a variable `i` to each element of the list 
2. Runs through the list sequentially using the variable `i` 
    

## 1.2 for with enumerate

1. The `enumerate()` function allows us to count the index, and hence link one or more lists in a for together, without using dictionaries. 
2. One must ensure that the lists being associated together, must have the same length - otherwise, the interpreter will run into an `ArrayIndexOutofLength` error. 
3. Here is a code-snippet of how it works: 

![](enumerate.png)

4. Though the default starting value is 0, one can always set the starting value using the following syntax:
```python
for i in enumerate(list_name, 10)
#the counter i will begin by assigning 10 to the first element of the list
```

## 1.3 for with range

1. The primary syntax for the `range()` function is as follows: 
```python
for i in range(start, end, step)
#the counter begins from index 'start', ends at index 'end-1', and continues each iteration by adding 'step' to itself
```

3. `range()` can also be used in place of `enumerate()`, by further using the length of the lists as the 'end' (hence point 2, of section 1.2 becomes really important!)

# 2 while

1. The syntax for this particular loop structure is as follows: 
```python
while(condition):
    #code block
```
Therefore, the loop keeps on repeating, till the condition equates to **True** - i.e. the condition is checked before the loop is executed. 

_a sidenote:_
There is another loop structure called the `do-while` loop, where the code block is 'done' before the condition is checked - this allows the loop to be execute at least once. Though it is beyond the scope of this section, it's a useful structure to keep in mind in certain cases.
