# 1 Interrupting the flow

There are two ways to sort of 'interrupt' the normal flow of iterative structures:
1. `break` - usually associated with a condition. If the condition is met, the loop stops iterating any further - i.e. it 'breaks.' 
2. `continue` - also associate with a condition. If the condition is met, the loop merely skips to the next iteration, without executing any of the coding statement after the `continue` statement - i.e. it 'continues' to the next iteration. 

These techniques can be used for both `for` and `while` loops.

# 2 List comprehension!

_essentially an optimisation of lists using for loops_

## 2.1 Basic syntax

1. The concept of list comprehension essentially creates a list using a for loop. 
2. The syntax of a basic list containing the first 5 numbers will be written as follows: 
```python
[number for number in range(1,6)]
```
3. Therefore, the `range(1,6)` is traversed using the variable `number` - the obtained number in each case is then stored in a list, by using `[]`. 
4. One can use basic mathematical calculations as well in the following way: 
```python
[number*2 for number in range(1,6)] #printing the double of every number between 1 and 6 (not inclusive of 6)
```

## 3.1 for with unpacking

1. Unpacking can be compared with subsetting - in the sense that it creates singleton subsets of each element of the list. 
2. It is of particular use when we are using multi-dimensional lists in a for-loop. 
3. The syntax is as follows: 
```python
list1=[[97, 98], [99, 100], [101, 102]]
for i, j in list1:
    print("num1," {i}, "and num2 ", {j})
#essentially the counters i and j traverse each row together, with i being used to single out the 0th element in each row, and j for the 1st element (by index) in each row

## 3.2 for with zip()

1. Much like the `.zip` extension on our computers, the `zip()` helps to combine two or more lists into a temporary single, multi-dimensional list, that can be traversed together. 
2. Here is a code snippet to explain how it works! 

![](zipfunc.png)

## 3.3 for with dictionaries

This is best explained directly using a code snippet, and one of the dictionaries we used earlier. 

![](dict_for.png)

## Footnotes
