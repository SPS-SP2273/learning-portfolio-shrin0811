```python
# Some data for plotting
from matplotlib import pyplot as plt
import numpy as np
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=0.1, yerr=3,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5, linewidth=2,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import math
x=np.linspace(0,50,200)
y=np.sin(x)*np.power(np.e, (-x/10))
plt.plot(x, y, linewidth=2) 
plt.grid(alpha=0.25)
plt.xlabel('x', color='darkorange', fontsize='15') 
plt.ylabel('y', color='darkorange', fontsize='15')
plt.title("y=$\\alpha^{-x/10}\sin(x)$", fontsize=20)
plt.ylim([-1,1])
```
