## Quadratic Equations

### Introduction

![](quadratic.png)

[Image from Wikipedia](https://en.wikipedia.org/wiki/Quadratic_function#/media/File:Polynomialdeg2.svg)

The general form of a quadratic equation is: 

$$
ax^2+bx+c=0
$$


![](https://en.wikipedia.org/wiki/Quadratic_equation#/media/File:Polynomialdeg2.svg)



### Solutions

Provided $a /ne 0$, we can use an elementary algebraic method called [Completing the square](https://en.wikipedia.org/wiki/Completing_the_square) to show that a quadratic equation has the following solution:

$$
x = \dfrac{-b \pm \sqrt{b^2-4ac}}{2a}
$$

### Discriminant

The quantity $\delta = {b^2-4ac}$ is called the **discriminant** of the equation and decides the nature of its solutions. The table below shows the various possibilities. 

|Discriminant|Roots|
|:--:|:--:|
|$\delta = {b^2-4ac} = 0$|A single solution of $\dfrac{-b}{2a}$|
|$\delta = {b^2-4ac} \gt 0$|Two distinct real solutions|
|$\delta = {b^2-4ac} \lt 0$|No real solutions - both are complex|

### Exercise 2
1. **Git repository**: A place where all the files and their changes for a project are stored.

2. Git keeps track of **changes** in the files in our repository. 

3. **Commit**: Saving your work in Git, like taking a snapshot of your project at a specific point.

4. **Push**: Sending your saved changes from your computer to the internet, so others can see and work with them.

5. **Pull**: Getting the latest changes from the internet to your computer, so you're up to date with what others have done.

6. **GitHub vs. GitHub Desktop**:

_GitHub_: An open-source website where you can save your projects and collaborate with others.

_GitHub Desktop_: An app that helps you manage your projects on GitHub using a simpler interface.
