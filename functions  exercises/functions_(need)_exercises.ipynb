Exercise 1

Essentially in Python, when we write an if statement followed by an else statement, the code inside the else block will execute only if the condition in the if statement evaluates to False. However, if there is no else block, the program will simply move on to the next block of code after the if statement if the condition is not met


```python
def calculate_bmi(weight, height):
    BMI=weight/(height ** 2)
    if(BMI<18.5):
        return "Underweight"
    elif(BMI<24.9):
        return "Normal weight"
    elif(BMI<29.9):
        return "Overweight"
    else:
        return "Obesity"

def main():
    w=float(input("Enter weight in kgs="))
    h=float(input("Enter height in ms="))
    print(calculate_bmi(w,h))

main()

```
