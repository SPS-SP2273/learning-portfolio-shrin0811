```python
def convert_celsius(temperature_celsius, target_scale):
    target_scale=target_scale.lower() 
    if(target_scale=='kelvin'):
        final_temp=temperature_celsius+273.15
    elif(target_scale=="fahrenheit"):
        final_temp=((9/5)*temperature_celsius)+32
    return final_temp
convert_celsius(float(input()), str(input()))
```

```python
def convert_fahrenheit(temperature_fahrenheit, target_scale):
    target_scale=target_scale.lower() 
    if(target_scale=='kelvin'):
        final_temp=((9/5)*temperature_fahrenheit)+32+273.15
    elif(target_scale=="celsius"):
        final_temp=((9/5)*temperature_fahrenheit)+32
    return final_temp
convert_fahrenheit(float(input()), str(input()))
```

```python
def convert_temperature(temperature, source_scale, target_Scale):
    source_scale=source_scale.lower()
    match source_scale:
        case "celsius":
            final_temp=convert_celsius(temperature, target_Scale)
        case "fahrenheit":
            final_temp=convert_fahrenheit(temperature, target_Scale)
        case "kelvin":
            if(target_Scale.lower()=="celsius"):
                final_temp=temperature-273.15
            elif(target_Scale.lower()=="fahrenheit"):
                final_temp=(9/5)*(temperature-273.15)+32
        case _:
            print("not applicable")
    print("converted temperature=", final_temp)

convert_temperature(float(input()), input(), input())

```
